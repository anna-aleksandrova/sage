```python
G = DihedralGroup(10)

g1 = G.list()[1]
g2 = G.list()[2]
print(g1, g2, sep='\n')
```

```python
H1 = G.subgroup([g1, g2])
H1
```

```python
H1.is_cyclic()
```

```python
H2 = G.subgroup([g1])
print(H2.list())
H2.is_cyclic()
```

```python
H2.cayley_graph().plot(layout='circular', vertex_labels=False)
```

```python
G.cayley_graph().plot()
```

```python
H1.structure_description()
```

```python
H1.is_isomorphic(CyclicPermutationGroup(5))
```
