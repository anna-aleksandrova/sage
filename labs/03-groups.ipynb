```python
from pprint import pprint
G = DihedralGroup(10)
subgroups = G.subgroups()

by_order = {}
for el in subgroups:
    n = el.cardinality()
    if n not in by_order:
        by_order[n] = []
    by_order[n].append(el)
pprint(by_order, indent=2)
```
