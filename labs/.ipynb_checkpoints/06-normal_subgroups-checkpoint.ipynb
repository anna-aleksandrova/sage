```python
G = DihedralGroup(10)
G.normal_subgroups()
```

```python
g1 = G('(1,3,5,7,9)(2,4,6,8,10)')
H = G.subgroup([g1])
Q = G.quotient(H)
Q.structure_description()
```

```python
for group in G.normal_subgroups():
    print(group.order())
```

```python
quotient = []
for g in G:
    cl = []
    for h in H:
        cl.append(g * h)
    cl.sort()
    if cl not in quotient:
        quotient.append(cl)
for el in quotient:
    print(el)
```

```python
left = G.cosets(H, side='left')
left
```

```python
right = G.cosets(H, side='right')
right
```

```python
a = G1('(1,2,3)')
H = G.subgroup([a])
tr = G.cosets(H, side='right')
for el in tr:
    el.sort()
print(tr)
tl = G.cosets(H, side='left')
for el in tl:
    el.sort()
print(tl)
```
