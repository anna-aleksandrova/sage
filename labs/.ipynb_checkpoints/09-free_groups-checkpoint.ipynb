```python
F1.<a, b> = FreeGroup()
G1 = F1 / [a*a, a*b*a*b]
F2.<x> = FreeGroup()
Z = F2 / []
G = G1.direct_product(Z)
G.rewriting_system()
```

```python
F1.<a, b> = FreeGroup()
G1 = F1 / [a*b*a*b^(-1)]
a = G1(a)
b = G1(b)
a*b*a == b
H = G.subgroup([a, b])
H.is_abelian()
```
