```python
# task1
"""
Comment: k is a splitting field for x^25 - x, thus the factorization of p(x) is not a suprise.
"""
k = GF(5^2, "alpha")
y = polygen(k, 'y')
P = y.parent()
p = P([0, -1]+[0 for i in range(23)] + [1])
p
```

```python
# task2
k.<a> = GF(2^7)
repr1 = k.list()
repr1
```

```python
a = k.gen()
repr2 = [a**i for i in range(1, 128)]
repr2.insert(0, k.list()[0])
repr2
```

```python
sorted(repr1)
sorted(repr2)
for i in range(128):
    if repr1[i] != repr2[i]:
        print(False)
        break
else:
    print(True)
```
