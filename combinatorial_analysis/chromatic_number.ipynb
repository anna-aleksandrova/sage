```python
def equal_degree_sequences(n):
    needed_set = list(graphs(n))
    classes = {}
    for i in range(len(needed_set)):
        for j in range(i + 1, len(needed_set)):
            temp1 = tuple(needed_set[i].degree_sequence())
            temp2 = tuple(needed_set[j].degree_sequence())
            if temp1 == temp2 and (needed_set[i].chromatic_number() != needed_set[j].chromatic_number()):
                if temp1 not in classes:
                    classes[temp1] = {(i, needed_set[i].chromatic_number()), (j, needed_set[j].chromatic_number())}
                else:
                    classes[temp1].add((j, needed_set[j].chromatic_number()))
    return classes, needed_set
equal_degree_sequences(5)[0]
```

```python
def present(cycles, cycle):
    for el in cycles:
        if set(el) == set(cycle):
            return True
    return False

def cycles_of_length3(G):
    cycles = []
    for i1 in range(6):
        for i2 in range(6):
            for i3 in range(6):
                current = (i1, i2, i3)
                if (G.has_edge(i1, i2) and G.has_edge(i2, i3) and G.has_edge(i3, i1) 
                    and (not present(cycles, current))):
                    cycles.append(current)
    return cycles
    
def cycles_of_length4(G):
    cycles = []
    for i1 in range(6):
        for i2 in range(6):
            for i3 in range(6):
                for i4 in range(6):
                    current = (i1, i2, i3, i4)
                    if (len(set(current)) == len(current) and G.has_edge(i1, i2) and 
                        G.has_edge(i2, i3) and G.has_edge(i3, i4) and 
                        G.has_edge(i4, i1) and (not present(cycles, current))):
                        cycles.append(current)
    return cycles

def cycles_of_length5(G):
    cycles = []
    for i1 in range(6):
        for i2 in range(6):
            for i3 in range(6):
                for i4 in range(6):
                    for i5 in range(6):
                        current = (i1, i2, i3, i4, i5)
                        if (len(set(current)) == len(current) and 
                            G.has_edge(i1, i2) and G.has_edge(i2, i3) and G.has_edge(i3, i4) and 
                            G.has_edge(i4, i5) and G.has_edge(i5, i1) and (not present(cycles, current))):
                            cycles.append(current)
    return cycles

def cycles_of_length6(G):
    cycles = []
    for i1 in range(6):
        for i2 in range(6):
            for i3 in range(6):
                for i4 in range(6):
                    for i5 in range(6):
                        for i6 in range(6):
                            current = (i1, i2, i3, i4, i5, i6)
                            if (len(set(current)) == len(current) and 
                                G.has_edge(i1, i2) and G.has_edge(i2, i3) and G.has_edge(i3, i4) and 
                                G.has_edge(i4, i5) and G.has_edge(i5, i6) and G.has_edge(i6, i1) and
                                (not present(cycles, current))):
                                cycles.append(current)
    return cycles
```
